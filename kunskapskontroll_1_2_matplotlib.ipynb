### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure is the whole figure and can be seen as a canvas
# Axes is the "canvas" which the plot is attached on. You can call methods that do the plotting from the axes, which can give a lot of flexibility to our plots.
# Axis sets the scale, limits genarate ticks (the marks on the axis) and ticklables (strings labeling the ticks).
# Artists is almost everything that's visable on the figure. Including figure, axes and axis. Most artists are tied to an axes.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("Probably the best plot in the world")

plt.figure()
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Probably the best plot in the world")

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create subplots
fig, ax = plt.subplots(1, 2, figsize = (8, 4), layout = 'constrained')

ax[0].scatter(random_data_x, random_data_y)
ax[0].set_title('Scatter Plot')

ax[1].bar(names, values)
ax[1].set_title('Bar Plot')
```
